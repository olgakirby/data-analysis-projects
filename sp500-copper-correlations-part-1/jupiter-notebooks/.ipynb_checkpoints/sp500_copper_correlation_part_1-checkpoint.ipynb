```python
#Check model assumptions for linear regression
#Create a plot of pairwise relationships between the variables

sns.pairplot(data)
```

```python
#Assumptions of linearity
#We can see a strong correlation between S&P500 and copper and Bitcoin and a slightly weaker correlation with 
#Crude Oil and Copper.
#For our analysis, we will assume S&P500 as our dependent variable and the otehr variables as our independent variables.
#We will run regression models to analyze how each of the independent variables correlates with the independent variable S&P500.
```

```python
#Model building: Copper - S&P500 
#Subset the necessary columns

sp500_copper = data[['sp500', 'copper']]
sp500_copper
```

```python
#Write the regression formula
ols_formula_copper = 'sp500 ~ copper'
```

```python
#Implement OLS
OLS_copper = ols(formula = ols_formula_copper, data = sp500_copper)
```

```python
#Fit the model to the data
model_copper = OLS_copper.fit()
```

```python
#Get a summary of the results
model_copper.summary()
```

```python
#Interpretation
```

```python
#Check the rest of model assumptions
#Plot the OLS data with the best fit regression line 
sns.regplot(x = 'copper', y = 'sp500', data = sp500_copper)
```

#Interpretation
The model suggests a strong positive relationship between copper prices and the S&P 500 index, with copper prices explaining a substantial portion of the variation in the S&P 500.
The F-statistic tests the overall significance of the model. A high F-statistic value of 575.1 indicates that the model is statistically significant.
The Prob (F-statistic) is extremely low (1.00e-67), meaning there is virtually no chance that the relationship observed in the sample is due to random chance.

```python
#Check model assumptions 
#Subset X variable 

X = sp500_copper[['copper']]
```

```python
# Get predictions from model

fitted_values = model_copper.predict(X)
```

```python
# Calculate residuals

residuals = model_copper.resid
```

```python
#Check the normality assumptions
import matplotlib.pyplot as plt

fig = sns.histplot(residuals)

fig.set_xlabel("Residual Values")

fig.set_title("Histogram of Residuals")

plt.show()
```

```python
#Create a qq plot 

import matplotlib.pyplot as plt

import statsmodels.api as sm

fig = sm.qqplot(model_copper.resid, line = 's')

plt.show()
```

#Interpretation
In this histogram, the distribution is not symmetric; the left side has more residuals concentrated in a narrower range, while the right side has fewer but more spread out residuals.

On the Q-Q plot the points deviate significantly from the line at both tails (both the lower left and upper right ends of the plot). This indicates that the residuals have heavier tails than a normal distribution, meaning they are not perfectly normally distributed. 

The histogram, like the Q-Q plot, suggests that the residuals are not normally distributed. The skewness and lack of symmetry indicate that the normality assumption for the residuals is likely violated.

```python
#Check for homoscedasticity assumption

fig = sns.scatterplot(x=fitted_values, y=residuals)

# Add reference line at residuals = 0

fig.axhline(0)

# Set x-axis and y-axis labels

fig.set_xlabel("Fitted Values")

fig.set_ylabel("Residuals")

# Show the plot

plt.show()
```

#Interpretation 

Homoscedasticity means that the variance of the residuals (errors) is constant across all levels of the independent variable(s) or fitted values.

The residuals appear to be more tightly clustered around the lower fitted values (left side of the plot) and spread out more as the fitted values increase (right side of the plot). This funnel-like shape is a common indicator of heteroscedasticity.

The presence of heteroscedasticity violates one of the key assumptions of ordinary least squares (OLS) regression, which assumes that residuals should have constant variance (homoscedasticity).

#Conclusion 

From the previous tests it is clear that the main conditions for the OLS regression analysis are violated. 

The reason for this could be that time series data often exhibit autocorrelation, where the value of a variable at one time point is correlated with its values at previous time points. This is particularly common in financial and economic data, such as copper prices and stock indices like the S&P 500.

This also violates the independence assumption of the model.

When the residuals are autocorrelated, the standard errors of the regression coefficients can be underestimated, leading to overly optimistic conclusions about the significance of predictors.

One common approach to mitigate autocorrelation is differencing the data, which involves modeling the changes in copper prices and the S&P 500 rather than their levels.
